```python
from torchvision import datasets
from torch.utils.data import DataLoader
from torchvision import transforms
from torchvision import models
import torch.utils.data as data
import torch
import torch.optim as optim
import numpy as np
from torch import optim, cuda
from torch.optim import lr_scheduler
import matplotlib.pyplot as plt
import os 
import torch.nn as nn
from google.colab import drive
drive.mount('/content/drive')
```

```python
batchSize = 16

train_on_gpu = cuda.is_available()
print(f'Train on gpu: {train_on_gpu}')

transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])
traindataPath = "/content/drive/My Drive/Data/TrainSet"

testdataPath = "/content/drive/My Drive/Data/TestSet"

TrainData = datasets.ImageFolder(root=traindataPath, transform=transform)
trainLoader = data.DataLoader(TrainData, shuffle=True, batch_size=batchSize)

TestData = datasets.ImageFolder(root=testdataPath, transform=transform)
testLoader =data.DataLoader(TestData, shuffle=True, batch_size=batchSize)

trainiter = iter(trainLoader)
features, labels = next(trainiter)
print(features.shape, labels.shape)
```

```python

model_ft = models.alexnet(pretrained=True)
print(model_ft)
for param in model_ft.parameters():
    param.requires_grad = False

n_inputs = model_ft.classifier[6].in_features


# Add on classifier
model_ft.classifier[6] = nn.Sequential(
                      nn.Linear(n_inputs, 256), 
                      nn.ReLU(), 
                      nn.Dropout(0.4),
                      nn.Linear(256, 8),                   
                      nn.LogSoftmax(dim=1))


print(model_ft)



if train_on_gpu:
    model_ft = model_ft.to('cuda')


net = model_ft

criterion = nn.MSELoss().cuda()
optimizer = optim.Adam(net.parameters(), lr = 0.0002, betas = (0.5, 0.999))

loss_list=[]
accuracy_list=[]


for epoch in range(150):  # loop over the dataset multiple times
    print(epoch)
    running_loss = 0.0
    total = 0
    correct = 0
    for i, data in enumerate(trainLoader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data
        # print(labels)

        target = torch.zeros(labels.size(0), 8)
        for i, row in enumerate(target, 0):
          row[labels[i]] = 1
        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs.cuda())
        loss = criterion(outputs, target.cuda())
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels.cuda()).sum().item()

    loss_list.append(running_loss)
    accuracy_list.append(100 * correct / total)

    print('[%d] loss: %.3f' %
                  (epoch + 1, running_loss / i))
    running_loss = 0.0
    total = 0
    correct = 0  
        
        # print statistics
        
            
torch.save(net, '/content/drive/My Drive/Data/alenet_model_modified_4.pt')   

print('Finished Training')
```

```python

fig,ax1 = plt.subplots()
color='tab:red'
ax1.plot(loss_list,color=color)
ax1.set_xlabel('epoch',color=color)
ax1.set_ylabel('Cost',color=color)
ax1.tick_params(axis='y',color=color)

ax2=ax1.twinx()
color='tab:blue'
ax2.set_ylabel('Accuracy',color=color)
ax2.set_xlabel('epoch',color=color)
ax2.plot(accuracy_list,color=color)
ax2.tick_params(axis='y',color=color)
fig.tight_layout()
```
