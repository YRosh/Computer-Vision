```python
import torch
import torch.nn as nn
from torchvision import transforms,datasets,models
import torch.utils.data as data
import pickle
import torch.optim as optim
```

```python
def weights_init(m):
  if type(m) in (nn.Linear, nn.Conv2d): 
    nn.init.xavier_uniform_(m.weight)

net = models.resnet18(pretrained=False)

def alter_model():
  numftrs = net.fc.in_features
  net.fc = nn.Linear(numftrs,8)

alter_model()
net.apply(weights_init)
net = net.cuda()
```

```python
batchSize = 32
# transform = transforms.Compose(
#     [transforms.ToTensor(),
#      transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])

transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

traindataPath = "/content/drive/My Drive/Colab Notebooks/Data/TrainSet"

testdataPath = "/content/drive/My Drive/Colab Notebooks/Data/TestSet"

TrainData = datasets.ImageFolder(root=traindataPath, transform=transform_train)
trainLoader = data.DataLoader(TrainData, shuffle=True, batch_size=batchSize)

TestData = datasets.ImageFolder(root=testdataPath, transform=transform_test)
testLoader = data.DataLoader(TestData, shuffle=True, batch_size=batchSize)
```

```python
criterion = nn.BCEWithLogitsLoss().cuda()
optimizer = optim.SGD(net.parameters(), lr=0.1,momentum=0.9, weight_decay=5e-4)
scheduler = torch.optim.lr_scheduler.MultiStepLR(optimizer, milestones=[150, 180], gamma=0.1, last_epoch=-1)
```

```python
loss_list=[]
accuracy_list=[]
for epoch in range(200):  # loop over the dataset multiple times
    running_loss = 0.0
    total=0
    correct=0
    for i, data in enumerate(trainLoader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data
        
        target = torch.zeros(labels.size(0), 8)
        for j, row in enumerate(target, 0):
          row[labels[j]] = 1
       
        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs.cuda())

        loss = criterion(outputs, target.cuda())
        loss.backward()
        optimizer.step()

        running_loss += loss.item()


        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels.cuda()).sum().item()

    scheduler.step()
    loss_list.append(running_loss)
    accuracy_list.append(100 * correct / total)
    print('[%d] loss: %.3f -- %.5f' % (epoch + 1, running_loss / i, optimizer.param_groups[0]['lr']))
    running_loss = 0.0  
    total=0
    correct=0

        
print('Finished Training')

torch.save(net, '/content/drive/My Drive/Colab Notebooks/Data/model_sch.pt')
```

```python
import matplotlib.pyplot as plt
fig,ax1 = plt.subplots()
color='tab:red'
ax1.plot(loss_list,color=color)
ax1.set_xlabel('epoch',color=color)
ax1.set_ylabel('Cost',color=color)
ax1.tick_params(axis='y',color=color)

ax2=ax1.twinx()
color='tab:blue'
ax2.set_ylabel('Accuracy',color=color)
ax2.set_xlabel('epoch',color=color)
ax2.plot(accuracy_list,color=color)
ax2.tick_params(axis='y',color=color)
fig.tight_layout()
```

```python
model = torch.load( '/content/drive/My Drive/Colab Notebooks/Data/model_sch.pt')
model = model.cuda()
model.eval()

correct = 0
total = 0
with torch.no_grad():
  for data in testLoader:
      images, labels = data

      outputs = model(images.cuda())
      _, predicted = torch.max(outputs.data, 1)
      total += labels.size(0)
      correct += (predicted == labels.cuda()).sum().item()

print('Accuracy of the network on the test images: %d %%' % (
    100 * correct / total))
```
